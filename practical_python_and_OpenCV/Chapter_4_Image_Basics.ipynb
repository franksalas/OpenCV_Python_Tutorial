# Image Basics

### Pixels
Images consist of rows and columns of pixels, they simple building blocks of an image.

ex: Image with a resolution of $500x300$ means it has 500 rows and 300 columns or $150,0000$ pixels

Piels are represented in two ways:

#### Grayscale
- each pixel has a value between 0 and 255, where 0 represents "black" and 255 represents "White". 
- The values between 0 and 255 are varying shades of gray, where values closer to 0 are darker and values closer to 255 are lighter.

#### Color
- Normally represented in the RGB color space. One calue for the Red component, one for Green, and one for Blue.Other colors exist but lets keep it simple.
- Each of these colors  is represented by an interger in the range 0 to 255, whic indicaltes how m"much" of the color ther is.
- Given that the pizel value only need to be in the range [0,255], we nomally use an 8-bit unsinge dint to represent each color intensity.
- We the n combine these value in to an RGB tuple in teh form (red, green, blue). This tuple represents our color.
- To construct a while color we would filll up each of the green, redm and blue buckend completely like this (255,255,255)
- To create a black color, we would emplty each bucket out: (0,0,0).

For reference, here are some common colors represented as RGB tuples:
- **Black**: (0,0,0)
- **White**:  (255,255,255)
- **Red **:  (255,0,0)
- **Blue **:  (0,0,255)
- **Aqua**:  (0,255,255)
- **Marron**:  (128,128,0)
- **Navy **:  (0,0,128)
- **Olive **:  (128,120,0)
- **Purple **:  (128,0,128)
- **Teal**:  (0,128,128)
- **Yellow **:  (255,255,0)





```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/trex.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
(r,g,b) = img[0,0]
print('Pixel at (0,0) - Red: {}, Green: {}, Blue: {}'.format(r,g,b))
```

```python
img[0,0] = (255,0,0)
(r,g,b) = img[0,0]
print('Pixel at (0,0) - Red: {}, Green: {}, Blue: {}'.format(r,g,b))
```

```python
corner = img[0:100, 0:100]
```

```python
# lets plot them
plt.figure(figsize=(15,10))

plt.subplot(2,2,1), plt.imshow(img,cmap='gray')
plt.title('Original'),plt.axis("off")

plt.subplot(2,2,2), plt.imshow(corner,cmap='gray')
plt.title('corner'),plt.axis("off")

img[0:100, 0:100] = (0,255,0)
plt.subplot(2,2,3), plt.imshow(img,cmap='gray')
plt.title('Updated'),plt.axis("off")

plt.show()
```
