```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/trex.png')
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
(b,g,r) = img[0,0]
print('Pixel at (0,0) - Blue: {}, Green: {}, Red: {}'.format(b,g,r))
```

```python
img[0,0] = (0,0,255)
(b,g,r) = img[0,0]
print('Pixel at (0,0) - Blue: {}, Green: {}, REd: {}'.format(b,g,r))
```

```python
corner = img[0:100, 0:100]
```

```python
# lets plot them
plt.figure(figsize=(15,10))

plt.subplot(2,2,1), plt.imshow(img,cmap='gray')
plt.title('Original'),plt.axis("off")

plt.subplot(2,2,2), plt.imshow(corner,cmap='gray')
plt.title('corner'),plt.axis("off")

img[0:100, 0:100] = (0,255,0)
plt.subplot(2,2,3), plt.imshow(img,cmap='gray')
plt.title('Updated'),plt.axis("off")

plt.show()
```
