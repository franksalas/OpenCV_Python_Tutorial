```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/trex.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()



# plt.figure(figsize=(20,10))

# plt.subplot(1,2,1), plt.imshow(img)
# plt.title('BGR'),plt.axis("off")

# plt.subplot(1,2,2), plt.imshow(img2)
# plt.title('RGB'),plt.axis("off")

# plt.show()
```

### Averaging

```python
average_blurred = np.hstack([
    cv2.blur(img, (3,3)),
    cv2.blur(img, (5,5)),
    cv2.blur(img, (7,7))])


plt.figure(figsize=(20,10))

plt.imshow(average_blurred)
plt.title("Averaged")
plt.axis("off")
plt.show()
```

### Gaussian

```python
gaussian_blurred = np.hstack([
    cv2.GaussianBlur(img, (3,3), 0),
    cv2.GaussianBlur(img, (5,5), 0),
    cv2.GaussianBlur(img, (7,7), 0)])


plt.figure(figsize=(20,10))

plt.imshow(gaussian_blurred)
plt.title("Gaussian")
plt.axis("off")
plt.show()
```

### Median

```python
median_blurred = np.hstack([
    cv2.medianBlur(img, 3),
    cv2.medianBlur(img, 5),
    cv2.medianBlur(img, 7)])


plt.figure(figsize=(20,10))

plt.imshow(median_blurred)
plt.title("Median")
plt.axis("off")
plt.show()
```

### Bilateral

```python
img = cv2.imread('images/beach.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

bilateral_blurred = np.hstack([
    cv2.bilateralFilter(img, 5, 21, 21),
    cv2.bilateralFilter(img, 7, 31, 31),
    cv2.bilateralFilter(img, 9, 41, 41)])


plt.figure(figsize=(20,10))

plt.imshow(bilateral_blurred)
plt.title("Bilateral")
plt.axis("off")
plt.show()
```
