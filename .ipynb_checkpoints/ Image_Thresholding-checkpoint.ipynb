### Image Thresholding

#### Simple Trhesholding
Here, the matter is straight foward. If pixel value is greater thatn a threshold value, it is assignend one value(may be white), els it is assigned another value (may be black).  The function used is  **cv2.threshold**. First argument is the source image, which **should be a grayscale image**. Second argumetn is the threshold vbalue which is used to classify the pixel values. Third agrumetn is the max Val which represent the value to be vient if puxel value is more than (somethime less than) the threshold value. Opencv provode sdirent sytle s of thesholodng an it si decided by the foutrh paramter of the function. Diffren types are:
- cv2.THRESH_BINARY
- cv2.THRESH_BINARY_INV
- cv2.THRESH_THRUNC
- cv2.TRESH_TOZERO
- cv2.TRESH_TOZERO_INV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/gradient.png',0)
img = cv2.medianBlur(img,5)

ret, tresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret, tresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret, tresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret, tresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret, tresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)




# lets plot them
plt.figure(figsize=(10,10))

plt.subplot(2,3,1), plt.imshow(img,cmap='gray')
plt.title('Original'),plt.axis("off")

plt.subplot(2,3,2), plt.imshow(tresh1,cmap='gray')
plt.title('BINARY'),plt.axis("off")

plt.subplot(2,3,3), plt.imshow(tresh2,cmap='gray')
plt.title('BINARY_INV'),plt.axis("off")

plt.subplot(2,3,4), plt.imshow(tresh3,cmap='gray')
plt.title('TRUNC'),plt.axis("off")

plt.subplot(2,3,5), plt.imshow(tresh4,cmap='gray')
plt.title('TOZERO'),plt.axis("off")

plt.subplot(2,3,6), plt.imshow(tresh5,cmap='gray')
plt.title('TOZERO_INV'),plt.axis("off")


plt.show()
```

### Adaptive Thresholding
In teh previous sections, we used a gloval value as threshold value. But it may not be good in all teh condition where images has diffrent light condition in diffrent areas. In that case wei go for adaptive thrsholindg . In this, the algorithmn calculate the threshold for a small region of the image. So we get diffrent thresholds for diffrent regions of the same image and it gives us betteer results for images with variayn illumiations.

It has threee 'special' input params and only one output.

**Adaptive method** It decides how shtresholding value is calculated.
- cv2.ADAPTIVE_THRESH_MEAN_C: threshold value is the mean of neighborhould area.
- cv2.ADAPTIVE_THRESH_GAUSSIAN_C: threshold value is the weighted sum  of neighborhood values where weight are a gaussian window.
**block size**: it decides the size of neighborhood area
**C**: it just a constan which is substractedfrom the mean or weigherd mean calculated

```python
img = cv2.imread('images/dave.jpg',0)
img = cv2.medianBlur(img,5)

ret, th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

# lets plot them
plt.figure(figsize=(15,15))

plt.subplot(2,2,1), plt.imshow(img,cmap='gray')
plt.title('Original'),plt.axis("off")

plt.subplot(2,2,2), plt.imshow(th1,cmap='gray')
plt.title('global Thresholding (v=127)'),plt.axis("off")

plt.subplot(2,2,3), plt.imshow(th2,cmap='gray')
plt.title('Adaptive Mean Thresholding'),plt.axis("off")

plt.subplot(2,2,4), plt.imshow(th3,cmap='gray')
plt.title('Adaptive Gaussian Thresholding'),plt.axis("off")


plt.show()
```
